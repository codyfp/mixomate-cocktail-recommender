```python
# Load your dataset
input_csv = "cocktail_data_mongo.csv"  # Update with your dataset path
output_csv = "updated_cocktail_data_mongo.csv"  # Update with the desired output path

# Read the CSV file into a DataFrame
df = pd.read_csv(input_csv)

# Create an empty list to store flavor profiles
flavor_profiles = []
```

```python
flavor_profiles=['Sour', 'Flowery', 'Fruity', 'Herby', 'Nutty', 'Carmelly', 'Chocolatey', 'Sweet', 'Spicy', 'Carbony', 'Pungent', 'Harsh', 'Sharp', 'Bitter', 'Mellow', 'Adicic', 'Winey']
```

```python
def generate_flavor_profiles(batch):
    # Define a list of messages for the chat conversation
    messages = [
        {"role": "system", "content": "You are a helpful assistant that recieves a list of ingredients and returns exactly 1 word, no more or less than 1 and no extra sentences."},
        {"role": "user", "content": f"The following is a list containing the ingredients of a specific cocktail drink. Choose a single word flavour profile from this list {flavor_profiles} (this aspect is important) for the cocktail based on its ingredients and which profile it most likey falls in and return the single word flavour profile (note: dont use N/A, assign every row a proper profile): {batch}"}
    ]

    # Generate a response from the assistant
    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=messages,
        max_tokens=1000,  # Adjust the maximum length of the response if needed
        request_timeout=15
    )

    # Extract and print the assistant's reply
    assistant_reply = response["choices"][0]["message"]["content"]
    return assistant_reply
```

```python
generated_flavor_profiles = []
```

```python
# Iterate through the ingredients column and batch them

for index, ingredients in enumerate(df['ingredients']):
    try: 
        profile = generate_flavor_profiles(ingredients)
    except:
        profile = 'NULL'
    print(index, profile)
    generated_flavor_profiles.append(profile)
    time.sleep(1.5)
```

```python
df['flavour_profile'] = generated_flavor_profiles
```

```python
# Save the DataFrame to a new CSV file
df.to_csv(output_csv, index=False)
```
